```
import numpy as np
import matplotlib.pyplot as plt
```

```
def cos(frek,tipe,ht,hr,d):
  #frek = frekuensi (MHz)
  #tipe = urban, suburban, open
  #ht = tinggi tower (meter)
  #hr = tinggi user (meter)
  #d = jarak (meter)
    if frek in range(150,1501):
        Ca=69.55
        Cb=26.16
    elif frek in range(1501,2001):
        Ca=46.3
        Cb=33.9
    
    if tipe=='urban' :
        Cm=0
    elif tipe=='suburban' :
        Cm= -2*((np.log10((frek/28)**2))-5.4)
    elif tipe=='open' :
        Cm= -4.78*(np.log10(frek))+18.33*(np.log10(frek))-40.94
    else :
        Cm='tipe salah'
    
    ahr = (1.1*np.log10(frek)-0.7)*hr-(1.56*np.log10(frek)-1.8)
    Lp = Ca+(Cb*np.log10(frek))-(13.83*np.log(ht))-ahr+((44.9-6.55*np.log10(ht))*np.log10(d))-Cm
    
    return Lp
```

```
tower=80  
user=1.5 
dmin=200 
dmax=700 
N=200 
d=np.random.randint(dmin,dmax,N)
```

```
def wilayah1():
  f=1800 
  loc='urban'
  LpUrban=cos(f,loc,tower,user,d)
  d_sort=d[np.argsort(d)]
  Urban_sort=LpUrban[np.argsort(d)]
  return d_sort,Urban_sort
```

```
def wilayah2():
  f=1800 
  loc='open'
  LpUrban=cos(f,loc,tower,user,d)
  d_sort=d[np.argsort(d)]
  Urban_sort=LpUrban[np.argsort(d)]
  return d_sort,Urban_sort
```

```
def wilayah3():
  f=1800 
  loc='suburban'
  LpUrban=cos(f,loc,tower,user,d)
  d_sort=d[np.argsort(d)]
  Urban_sort=LpUrban[np.argsort(d)]
  return d_sort,Urban_sort
```

```
def frekuensi1():
  f=700 
  loc='urban'
  LpUrban=cos(f,loc,tower,user,d)
  d_sort=d[np.argsort(d)]
  Urban_sort=LpUrban[np.argsort(d)]
  return d_sort,Urban_sort
```

```
def frekuensi2():
  f=900 
  loc='urban'
  LpUrban=cos(f,loc,tower,user,d)
  d_sort=d[np.argsort(d)]
  Urban_sort=LpUrban[np.argsort(d)]
  return d_sort,Urban_sort
```

```
def frekuensi3():
  f=1800 
  loc='urban'
  LpUrban=cos(f,loc,tower,user,d)
  d_sort=d[np.argsort(d)]
  Urban_sort=LpUrban[np.argsort(d)]
  return d_sort,Urban_sort
```

```
q1,w1 = wilayah1()
q2,w2 = wilayah2()
q3,w3 = wilayah3()

z1,x1 = frekuensi1()
z2,x2 = frekuensi2()
z3,x3 = frekuensi3()
```

```
plt.figure()
plt.plot(q1,w1,'k', label='urban')
plt.plot(q2,w2,'y', label='open')
plt.plot(q3,w3,'c', label='suburban')
plt.legend()
plt.xlabel('Jarak (m)')
plt.ylabel('Path Loss (dB)')
plt.show()

plt.figure()
plt.plot(z1,x1,'k', label='700 Mhz')
plt.plot(z2,x2,'y', label='900 Mhz')
plt.plot(z3,x3,'c', label='1800 Mhz')
plt.legend()
plt.xlabel('Jarak (m)')
plt.ylabel('Path Loss (dB)')
plt.show()
```
